```python
def convert_to_years(row):
    if pd.notna(row) and isinstance(row, str):
        if 'years' in row:
            parts = row.split()
            years = float(parts[0])
            months = float(parts[2]) if len(parts) > 2 else 0
            return years + months / 12
    return row



# # Function to convert 'remaining_lease' to a numeric value
# def lease_to_years(lease_str):
#     if pd.isna(lease_str):
#         return np.nan
#     years, months = map(int, lease_str.split(' ')[::2])
#     return years + months / 12

# # Apply the conversion function
# df['remaining_lease_years'] = df['remaining_lease'].apply(lease_to_years)
```

```python
import matplotlib.pyplot as plt

# Columns to process
columns = ['block', 'floor_area_sqm', 'resale_price', 'remaining_lease', 
           'lease_commence_date', 'storey_lower_bound', 'storey_upper_bound']

for col_name in columns:
    col = df[col_name]
    
    # Create subplots
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Plot before removing outliers
    axs[0].boxplot(col)
    axs[0].set_title(f'Boxplot of {col_name} (Before Removing Outliers)')

    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f'Lower bound of {col_name}:', lower_bound)
    print(f'Upper bound of {col_name}:', upper_bound)

    # Clipping the column in the DataFrame
    df[col_name] = col.clip(lower=lower_bound, upper=upper_bound)

    # Plot after removing outliers
    axs[1].boxplot(df[col_name])
    axs[1].set_title(f'Boxplot of {col_name} (After Removing Outliers)')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create and train the Random Forest Regressor model
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_regressor.predict(X_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared (R2) Score
rf_r2 = r2_score(y_test, y_pred)

# mean absolute percentage error
mape_rf = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("Mean Absolute Error (MAE)    :", mae)
print("Mean Squared Error (MSE)     :", mse)
print("Root Mean Square Error (RMSE):", rmse)
print("R-squared (R2) Score         :", rf_r2)
print(f"Mean Absolute Percentage Error: {mape_rf:.2f}%")
```

finalizing the models

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
models = ['KNeighborsRegressor', 'DecisionTreeRegressor', 'RandomForestRegressor', 'XGBRegressor','linear_model']
score = [r2_knn, r2_dt, rf_r2, xgb_r2,r2_lr]

# Create a DataFrame
data = pd.DataFrame({
    'Model': models,
    'R-squared': score
})

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='R-squared', data=data, palette='viridis')
plt.title('R-squared of Regression Models')
plt.xticks(rotation=45)
plt.show()

# Finding the best fit model
print("FINDING BEST FIT MODEL FOR EVALUATION BASED ON R-squared SCORE\n")
max_score = max(score)  # Initialize max_score with the maximum value in the score list
max_model = ""

for model, score_value in zip(models, score):
    print(model, score_value)
    if score_value == max_score:
        max_model = model

print(f"\nModel: {max_model} contains Maximum R-squared Score: {max_score}")
```
